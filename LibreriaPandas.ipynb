```python
import pandas as pd #importamos libreria pandas
```

```python
import pickle #importamos modulo pickle
```

```python
with open('subjects.pkl', 'rb') as file:
        subjects = pickle.load(file) #importamos archivo subjects.pkl al notebook
```

```python
with open('officers.pkl', 'rb') as file:
        officers = pickle.load(file) #importamos archivo officers.pkl al notebook
```

```python
df_subjects = pd.DataFrame(subjects) #creamos el DataFrame df_subjects
```

```python
df_officers = pd.DataFrame(officers) #creamos el DataFrame df_officers
```

```python
df1 = pd.read_csv("Cleaned_DS_Jobs.csv", on_bad_lines="skip") #importamos archivo csv y lo guardamos en un DataFrame
```

```python
df1 #mostramos el DataFrame
```

```python
lista_1 = ["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"] #Creamos una lista
```

```python
df2 = pd.DataFrame(lista_1) #guardamos los valores de la lista en un nuevo DataFrame
```

```python
df3 = df1.replace (("na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"), np.nan) #reemplazamos los valores de cada elemento del DataFrame por "NaN"
```

```python
df3 #mostramos el DataFrame resultante
```

```python
df4 = df3.dropna(axis=0) #eliminamos filas que contienen valores NaN
```

```python
df4 #mostramos el DataFrame resultante
```

```python
reemplazos = {
    "10000+ employees": "Mega Empresas",
    "5001 to 10000 employees": "Grandes Empresas",
    "1001 to 5000 employees": "Medianas Empresas",
    "201 to 500 employees": "Pequeñas Empresas",
    "51 to 200 employees": "Pequeñas Grandes Empresas",
    "501 to 1000 employees": "Microempresas",
    "Unknown": "Empresas sin Información"
} #creamos un diccionario con strings a reemplazar
```

```python
df4["Size"] = df4["Size"].replace(reemplazos) #renombranos los valores de la columna size del df4 por los valores establecidos en el nuevo diccionario
```

```python
df4["Size"].unique() #mostramos los nuevos valores de la columna Size del df4
```

```python
df4 #mostramos el DataFrame
```

```python
df4[["Salary Estimate min", "Salary Estimate max"]] = df4["Salary Estimate"].str.split("-", expand = True) #dividimos la columna Salary Estimate y generamos dos columnas, la primera con el salario estimado minimo y la segunda con el salario estimado maximo
```

```python
df4 #mostramos el dataframe resultante
```

```python
df4["Salary Estimate min"] = pd.to_numeric(df4["Salary Estimate min"], errors="coerce") #convertimos valores de string a numericos en la columna Salary Estimate min
```

```python
df_pivot_reporte2 = pd.pivot_table (
    data = df4
    , index = "Size"
    , values = "Salary Estimate min"
    , aggfunc = "mean"
    , margins = True) #generamos un tabla pivote, el indice seria el tamaño de la empresa y calcularemos el promedio de la columna Salary Estimate min
```

```python
df_pivot_reporte2 #mostramos el DataFrame resultante
```

```python
df4["Salary Estimate max"] = pd.to_numeric(df4["Salary Estimate max"], errors="coerce") #convertimos valores de string a numericos en la columna Salary Estimate max
```

```python
df_pivot_reporte3 = pd.pivot_table (
    data = df4
    , index = "Size"
    , values = "Salary Estimate max"
    , aggfunc = "mean"
    , margins = True) #generamos un tabla pivote, el indice seria el tamaño de la empresa y calcularemos el promedio de la columna Salary Estimate max
```

```python
df_pivot_reporte3 #mostramos el DataFrame resultante
```
