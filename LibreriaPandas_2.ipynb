```python
import pandas as pd #importamos la libreria pandas con el alias pd
```

```python
df_resumen_resultados_astronautas = pd.read_csv('resumen_resultados_astronautas.csv', encoding='utf-8') #importamos el archivo y creamos el dataframe
```

```python
df_resumen_resultados_astronautas #mostramos los datos del dataframe
```

```python
df_resumen_resultados_astronautas.head() #mostramos las primeras 5 filas del dataframe
```

```python
df_resumen_resultados_astronautas["altura_m"] = df_resumen_resultados_astronautas["altura"] / 100 #realizamos conversión de altura de centimetros a metros
```

```python
df_resumen_resultados_astronautas["IMC"] = df_resumen_resultados_astronautas["peso"] / df_resumen_resultados_astronautas["altura_m"] ** 2 #creamos otra nueva columna llamada IMC y almacenamos la formula
```

```python
df_resumen_resultados_astronautas #mostramos el dataframe con las nuevas columnas para validar resultados
```

```python
df_resumen_resultados_astronautas['promedio_evaluaciones'] = df_resumen_resultados_astronautas[[
    'evaluacion_fisica',
    'evaluacion_psicologica',
    'evaluacion_adaptativa',
    'evaluacion_medica'
]].mean(axis=1) #calculamos el promedio de las 4 evaluaciones por astronauta
```

```python
df_resumen_resultados_astronautas.head() #mostramos el dataframe nuevamente para validar resultados de la nueva columna
```

```python
df_IMC_adecuado = df_resumen_resultados_astronautas[(df_resumen_resultados_astronautas["IMC"] >= 18.5) &
                                                    (df_resumen_resultados_astronautas["IMC"] <= 24.9)] #aplicamos filtro de IMC entre los limites adecuados
```

```python
df_IMC_adecuado = df_resumen_resultados_astronautas[df_resumen_resultados_astronautas['IMC_adecuado']] #guardamos IMC adecuado en la variable del dataframe
```

```python
print(df_IMC_adecuado[["nombre", "IMC"]]) #imprimimos nombres de candidatos que tienen IMC adecuado de acuerdo a rangos establecidos
```

```python
df_filtrado = df_resumen_resultados_astronautas[
    (df_resumen_resultados_astronautas["IMC"] >= 18.5) &
    (df_resumen_resultados_astronautas["IMC"] <= 24.9) &
    (df_resumen_resultados_astronautas['promedio_evaluaciones'] >= 87)
] #agregamos una nueva condicion de filtrado (promedio de evaluaciones debe ser mayor o igual a 87)
```

```python
print(df_filtrado[['nombre', 'IMC', 'promedio_evaluaciones']]) #imprimimos las columnas nombre, IMC y promedio_evaluaciones para revisar que se cumplan las condiciones establecidas
```

```python
df_filtrado.reset_index(drop=True, inplace=True) #reiniciamos el indice de dataframe resultante
```

```python
print(df_filtrado[['nombre', 'IMC', 'promedio_evaluaciones']]) #imprimimos nuevamente las columnas
```

```python
df_filtrado['califica'] = 'Si' #modificamos para que la columna califica diga si en todas las columnas del dataframe resultante
```

```python
print(df_filtrado[['nombre', 'IMC', 'promedio_evaluaciones', 'califica']]) #imprimos resultados
```

```python
df_filtrado.to_csv('astronautas_calificados.csv', index=False) #exportamos datos a un archivo csv
```
