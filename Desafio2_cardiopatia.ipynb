```python
categoricas = ["cp", "restecg", "slope", "thal", "sex", "fbs", "exang"]

# Reemplazar ? por NaN
df_final = df_final.replace('?', np.nan)

# Convertir a numérico
for col in df_final.columns:
    df_final[col] = pd.to_numeric(df_final[col], errors='coerce')

# Eliminar columnas con >25% de NaN
df_final = df_final.loc[:, df_final.isnull().mean() <= 0.25]

# Rellenar valores ausentes por clase
for col in df_final.columns:
    if col != 'num' and df_final[col].isnull().sum() > 0:
        if col in categoricas:
            df_final[col] = df_final.groupby('num')[col].transform(
                lambda x: x.fillna(round(x.mean()))
            )
        else:
            df_final[col] = df_final.groupby('num')[col].transform(
                lambda x: x.fillna(x.mean())
            )
```

### MODELAMIENTO Y MÉTRICAS DE EVALUACIÓN

```python
X = df_final[["age", "sex", "cp", "threstbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak"]]
y = df_final["num"]
```

```python
# Dividir los datos en entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33, random_state=23)
```

### Módelo de ensamble  AdaBoost

```python
%%time
# aprendiz débil para el ensamble
learner_base = DecisionTreeClassifier(max_depth=2)

# creamos el modelo de ensamble AdaBoost con 25 iteraciones
aboost_m1 = AdaBoostClassifier(estimator=learner_base, n_estimators=25, learning_rate=.75)

# entrenamos el modelo con los datos de entrenamiento
aboost_m1.fit(X_train, y_train)

# obtener predicciones para el conjunto de test
yhat_test = aboost_m1.predict(X_test)

print(yhat_test)
```

```python
yhat_train = aboost_m1.predict(X_train)
err = 1 - accuracy_score(y_train, yhat_train)
print(err)

print(classification_report(y_train, yhat_train))
```

```python
yhat_test = aboost_m1.predict(X_test)
err = 1 - accuracy_score(y_test, yhat_test)
print(err)

print(classification_report(y_test, yhat_test))
```

##### De lo anterior, podemos observar que las métricas para el conjunto de entrenamiento son mucho mejores que para el conjunto de prueba debido a que tenemos una exactitud del 93% para train vs una exactitud del 77% para test; así como un f1-score promediado con los mismos porcentajes para cada caso.

```python
# Obtener las probabilidades predichas para la clase positiva (índice 1)
yhat_test_proba = aboost_m1.predict_proba(X_test)[:, 1]
yhat_train_proba = aboost_m1.predict_proba(X_train)[:, 1]

# Calcular la tasa de verdaderos positivos (TPR) y la tasa de falsos positivos (FPR)
fpr_test, tpr_test, umbrales_test = roc_curve(y_test, yhat_test_proba)
fpr_train, tpr_train, umbrales_train = roc_curve(y_train, yhat_train_proba)

# Calcular el área bajo la curva ROC (AUC)
auc_test = roc_auc_score(y_test, yhat_test_proba)
auc_train = roc_auc_score(y_train, yhat_train_proba)

# Plot de la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr_test, tpr_test, lw=1.5, label='Curva ROC test (AUC = {:.2f})'.format(auc_test))
plt.plot(fpr_train, tpr_train, lw=1.5, label='Curva ROC train (AUC = {:.2f})'.format(auc_train))

plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=1)
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - AdaBoost')
plt.legend(loc='lower right')
plt.show()
```

##### Para la curva ROC del modelo AdaBoost, se observan datos tendientes a la presencia de overfitting debido a que AUC para train es mucho mayor que AUC para test. Además el modelo clasifica casí que perfecto en train.

```python
from xgboost import XGBClassifier

# Definir la grilla de hiperparámetros
param_grid = {
    'n_estimators': np.linspace(20, 200, 15, dtype=int),
    'learning_rate': np.linspace(0.004, 1.0, 40),
    'subsample': np.linspace(0.1, 1.0, 18)
}

# Inicializar el modelo
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# Configurar la búsqueda con GridSearchCV usando f1 como métrica
grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    scoring='f1',
    cv=5,
    verbose=2,
    n_jobs=-1
)

# Entrenar la búsqueda
grid_search.fit(X_train, y_train)

# Mostrar los mejores hiperparámetros
print("Mejores hiperparámetros:", grid_search.best_params_)
print("Mejor F1-score en validación:", grid_search.best_score_)

# Entrenar el modelo final con los mejores hiperparámetros
best_xgb = grid_search.best_estimator_
y_pred = best_xgb.predict(X_test)

# Mostrar métricas
print("\nMétricas en test:")
print(classification_report(y_test, y_pred))
```

```python
modelos = {
    'AdaBoost': aboost_m1,
    'XGBoost': xgb_model,
    'GradBoost': gboost_m1
}

resultados = []
for nombre, m in modelos.items():
    y_pred = m.predict(X_test)
    # Probabilidades (si el modelo las soporta)
    if hasattr(m, "predict_proba"):
        y_proba = m.predict_proba(X_test)[:, 1]
        auc = roc_auc_score(y_test, y_proba)
    else:
        auc = float('nan')
    resultados.append({
        'Modelo': nombre,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, zero_division=0),
        'Recall': recall_score(y_test, y_pred, zero_division=0),
        'F1': f1_score(y_test, y_pred, zero_division=0),
        'AUC': auc
    })

df_resultados = pd.DataFrame(resultados).sort_values(by='F1', ascending=False).reset_index(drop=True)
print(df_resultados)
```
